```python
def extract_category(base_url):
    response = requests.get(base_url)
    soup = BeautifulSoup(response.content)
    category_list = soup.find('ul',class_="nav nav-list").ul.find_all('li')

    for li in category_list:

        categories_dict[li.a.text.strip()] = base_url + li.a['href']
    return categories_dict

extract_category(base_url)
```

```python
cat_url = 'https://books.toscrape.com/catalogue/category/books/mystery_3/index.html'
```

```python
resp = requests.get(cat_url)
resp
```

```python
soup = BeautifulSoup(resp.content)
book_page_list = soup.find('ol',class_="row").find_all('li')
len(book_page_list)
```

```python
book_page_list[0].a['href']
```

```python
book_suffix_link = book_page_list[0].a['href'].split('/')[3:]
book_suffix_link = "/".join(book_suffix_link)
book_suffix_link
```

```python
base_url +'catalogue/'+book_suffix_link
```

```python
data_dict = {'prod_page_url' : [],
            'title' : [],
             'price_inc_tax' : [],
             'num_availability' : []}

for book_1 in book_page_list :
    book_suffix_link = book_1.a['href'].split('/')[3:]
    book_suffix_link = "/".join(book_suffix_link)
    book_url = base_url +'catalogue/'+book_suffix_link
    
    data_dict['prod_page_url'].append(book_url)
    
    # getting price and availabolity from table tag
    resp =requests.get(book_url)
    soup = BeautifulSoup(resp.content)
    print(book_url)
    
    rows = soup.find('table',class_='table table-striped')
    price = rows.find_all('tr')[3].td.text.strip()
    Availability = rows.find_all('tr')[5].td.text.strip()
    
    #getting book_name from ul tag 
    book_name = soup.find('ul',class_='breadcrumb').find('li',class_='active').text.strip()
    
    data_dict['title'].append(book_name)
    data_dict['price_inc_tax'].append(price)
    data_dict['num_availability'].append(Availability)
                              
```
